```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# s: 辺の情報を表す文字列
# n: ノードの数
# m: 辺の数
def viz_graph(s, n, m):
    inp = list(filter(lambda x: x != "", s.split('\n')))
    G = nx.Graph()
    for i in range(n): G.add_node(i)
    for i in range(m):
        u, v = map(int, inp[i].split())
        u-=1
        v-=1
        G.add_edge(u, v)
    nx.draw_shell(G, with_labels=True)
```

```python
s = """
1 2
1 3
2 4
4 5
4 6
3 7
7 8
8 9
8 10
"""

viz_graph(s, 10, 9)
```

```python
# s: 辺の情報を表す文字列
# n: ノードの数
# m: 辺の数
def viz_digraph(s, n, m):
    inp = list(filter(lambda x: x != "", s.split('\n')))
    G = nx.DiGraph()
    for i in range(n): G.add_node(i)
    for i in range(m):
        u, v = map(int, inp[i].split())
        u-=1
        v-=1
        G.add_edge(u, v)
    nx.draw_shell(G, with_labels=True)
```

```python
s = """
1 2
2 1
2 3
4 3
4 1
1 4
2 3

"""

viz_digraph(s, 4, 7)
```
